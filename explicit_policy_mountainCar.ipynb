```python
import numpy as np
import gym
import matplotlib.pyplot as plt
```

```python
def policy(velocity):
    # 0 - backward, 1 - none, 2 - forward
    if velocity < 0:
        return 0
    elif velocity >= 0:
        return 2
```

```python
if __name__ == '__main__':
    env = gym.make('MountainCar-v0')

    numGames = 100
    rewards = np.zeros(numGames)
    for i in range(numGames):
        observation = env.reset()
        done = False
        epRewards = 0
        while not done:
            action = policy(observation[1])
            observation, reward, done, info = env.step(action)
            epRewards += reward
            # env.render()
        rewards[i] = epRewards

    plt.plot(rewards)
    plt.show()
```
