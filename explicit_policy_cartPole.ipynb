```python
import numpy as np
import matplotlib.pyplot as plt
import gym
from gym import wrappers
```

```python
def explicit_policy(state):
    action = 0 if state < 0 else 1   # If state is neg then return 0 else return 1
    return action

env = gym.make('CartPole-v0')
```

```python
totalRewards = []
for i in range(1000):
    observation = env.reset()
    epRewards = 0    
    done = False
    while not done:      
        action = explicit_policy(observation[2])            
        observation_, reward, done, info = env.step(action)
        epRewards += reward
        observation = observation_
    totalRewards.append(epRewards)

# env.render()
plt.plot(totalRewards)
plt.show() 
```
